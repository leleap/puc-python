#### 5. Complete o código acima usando estruturas condicionais (`if`, `elif`, `else`) para mostrar a classificação do IMC.

#### Regras

| IMC | Classificação |
|---|---|
| menor que 18.5 | Abaixo do peso |
| entre 18.5 e 24.9 | Peso normal |
| entre 25 e 29.9 | Sobrepeso |
| maior ou igual a 30 | Obesidade |

````python
#QUESTÃO 5

import time

peso_texto = input("Digite o seu peso (kg): ")
peso_texto = peso_texto.replace(',', '.')
peso = float(peso_texto)


print("Peso informado:", peso, "kg")

altura_texto = input("Digite a sua altura (m): ")
altura_texto = altura_texto.replace(',', '.')
altura = float(altura_texto)

print(f"Altura informada: {altura} m")

print("\nCalculando o IMC...")
time.sleep(2)

altura_ao_quadrado = altura ** 2
imc = peso / altura_ao_quadrado

if imc < 18.5:
  classificacao = "abaixo do peso"

elif imc <= 24.9:
  classificacao = "peso normal"

elif imc <= 29.9:
  classificacao = "sobrepeso"

else:
  classificacao = "obesidade"

print(f"\nSeu IMC é: {imc:.2f}")
print(f"Classificacao: {classificacao}")
````

#### 6. Adicione tratamento de erros usando `try` e `except` no código acima

O programa deve:

- impedir que o usuário digite letras no peso ou altura
- impedir divisão por zero
- mostrar mensagens de erro amigáveis

#### Regras

- Use `try`
- Use `except ValueError`
- Use `except ZeroDivisionError`

````python
import time

#Entrada do peso.
try:
    peso_texto = input("Digite o seu peso (kg): ")
    peso_texto = peso_texto.replace(',', '.')
    peso = float(peso_texto)

    if peso <= 0:
        raise ValueError("O peso deve ser maior que zero.")

#Entrada da altura.
    print("Peso informado:", peso, "kg")
    altura_texto = input("Digite a sua altura (m): ")
    altura_texto = altura_texto.replace(',', '.')
    altura = float(altura_texto)

    if altura <= 0:
        raise ValueError("A altura deve ser maior que zero.")

    print(f"Altura informada: {altura} m")

    print("\nCalculando o IMC...")
    time.sleep(2)

#Calcula o IMC.
    altura_ao_quadrado = altura ** 2
    imc = peso / altura_ao_quadrado

    print(f"\nSeu IMC é: {imc:.2f}")


except ValueError as erro:
    print(f"\nErro: {erro}")
    print("Por favor, insira valores numéricos válidos para o peso e a altura.")

except ZeroDivisionError:
    print(f"\nErro: A altura não pode ser zero.")

except Exception as erro:
    print(f"\nErro: {erro}")
    print("Ocorreu um erro inesperado. Por favor, tente novamente.")
````

#### 7. Modifique o programa para usar `while True`.

O programa deve continuar calculando IMC até o usuário digitar:

```text
sair

````python
import time

while True:
    try:
        # Entrada do peso
        peso_texto = input("Digite o seu peso (kg) ou 'sair': ")

        if peso_texto.lower() == "sair":
            print("Programa encerrado.")
            break

        peso_texto = peso_texto.replace(',', '.')
        peso = float(peso_texto)

        if peso <= 0:
            raise ValueError("O peso deve ser maior que zero.")

        print("Peso informado:", peso, "kg")

        # Entrada da altura
        altura_texto = input("Digite a sua altura (m) ou 'sair': ")

        if altura_texto.lower() == "sair":
            print("Programa encerrado.")
            break

        altura_texto = altura_texto.replace(',', '.')
        altura = float(altura_texto)

        if altura <= 0:
            raise ValueError("A altura deve ser maior que zero.")

        print(f"Altura informada: {altura} m")

        print("\nCalculando o IMC...")
        time.sleep(2)

        # Calcula o IMC
        altura_ao_quadrado = altura ** 2
        imc = peso / altura_ao_quadrado

        print(f"\nSeu IMC é: {imc:.2f}\n")

    except ValueError as erro:
        print(f"\nErro: {erro}")
        print("Por favor, insira valores numéricos válidos.\n")

    except ZeroDivisionError:
        print("\nErro: A altura não pode ser zero.\n")

    except Exception as erro:
        print(f"\nErro inesperado: {erro}\n")
````
